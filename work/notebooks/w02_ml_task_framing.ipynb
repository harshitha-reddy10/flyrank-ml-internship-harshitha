## My lane as an ML task (type)

My chosen lane is **Content Performance Prediction**.

This is a **classification** task because the goal is to classify whether a content page is likely to decline in performance or not. The model predicts whether a page belongs to the "declining" or "not declining" category, helping identify pages that should be prioritized for refresh.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

## Target or proxy

The target I want to predict is whether a content page is **declining** or **not declining**. I will use the `trend_direction` column as the target label, where pages marked as **"down"** represent declining content. This is an **observed outcome** from the dataset rather than a manually created rule, making it suitable for training and evaluating a classification model.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## Success metric

The main success metric is **Precision@50**. This measures how many of the top 50 pages selected by the model are actually declining. A high Precision@50 means the model is effectively identifying the pages that should be prioritized for content refresh, helping the content team focus on the most important pages first.

## The unit of analysis

The unit of analysis is **one content page**. Each row in the dataset represents a single content page with its search performance, content characteristics, and engagement metrics. The model uses these features to predict whether that page is likely to be declining.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why ML beats a fixed rule here

A fixed rule relies on simple conditions and cannot capture complex relationships between multiple features. A machine learning model can learn patterns from search performance, content characteristics, and engagement metrics together, allowing it to make more accurate predictions about declining content. This helps the content team prioritize pages more effectively than using a single predefined rule.
